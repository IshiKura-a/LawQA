```python
# coding=utf-8
from typing import List, Dict
import codecs, ast, random
import torch
file_path = "/data/home/ganleilei/law/law-qa/qa-18w.csv"
SEED=1024
```

```python
class Item():
    def __init__(self, line: str):
        super().__init__()
        parts = line.split('\t')
        self.dataid = parts[19]
        self.guid = parts[21]
        self.url = parts[0]
        self.pt = parts[1]
        self.status = parts[5]
        self.category = parts[11]
        self.category1 = parts[2]
        self.category2 = parts[7]
        self.datetime = parts[4]
        self.city = parts[6]
        self.web = parts[8]
        self.question_title = parts[9]
        self.question_content = parts[10]
        self.question_content_mask = parts[-6]
        self.question_content_cut_all_string = parts[-8]
        self.question_content_cut_string_stopwords = parts[-7]
        self.answers = ast.literal_eval(parts[3])
        self.answers_mask = parts[12]
        self.answers_mask_cut_string = parts[13]
        self.answers_can_show = parts[-3]
        self.domain = parts[-5]
        self.domain_v0117 = parts[-1]
```

```python
rf = codecs.open(file_path, mode='r', encoding='utf-8')
lines = rf.readlines()
```

```python
random.seed(SEED)
sample_num = 200
print("line number:", len(lines))
print(lines[1])
line_ids = [random.randint(1, len(lines)) for _ in range(sample_num)]
illegal_line_num = 0
ave_answer_num = 0
ave_question_len = 0
ave_answer_len = 0
contract_num, divorce_num = 0, 0
for line_id, line in enumerate(lines[1:2000]):
    if len(line.split('\t')) != 23: 
        illegal_line_num += 1
        continue
    # if line_id not in line_ids: continue
    item = Item(line)
    if '租房' in item.question_content and '合同' in item.question_content: 
        contract_num += 1
        print(item.url)
        print(item.question_content)
        print(item.category)
        print(item.category1)
        print(item.category2)
        print(item.answers)
    
    if '婚姻' in item.question_content:
        divorce_num += 1
        print(item.question_content)
        print(item.answers)

print("illegal line num:", illegal_line_num)
print("contract num:", contract_num)
```
